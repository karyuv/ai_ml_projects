```python
def load_dataframe_from_csv(csv_file_name:str)->pd.DataFrame: 
    """Load a csv file and return a pandas dataframe.

    Parameters
    ----------
    csv_file_name : str
        The csv file

    Return
    ------
    data_frame : DataFrame
        Pandas data frame

    Raises
    ------
    AssertionError
        If the file type is not of type csv
    """

    file_name = str(csv_file_name)
    _,file_type=os.path.splitext(file_name)
    assert file_type == '.csv',f"File format should be of type csv instead of format {file_type}"

    data_frame = pd.read_csv(csv_file_name)
    return data_frame

def print_frame_meta_info(data_frame:pd.DataFrame): 
    """Load a data frame file and print meta info of the data frame.

    Parameters
    ----------
    data_frame : DataFrame
    """
    rows,columns = np.shape(data_frame)
    print(f"Shape of the data frame is: {rows},{columns}")
    print(f"Number of variables are: {columns}")
    print("Memory Usage:"+str(data_frame.memory_usage(index=False,deep=True).sum()*0.001))
    print("")
    print("")

    
    for column in data_frame.columns:
        print(f"Name of the column: {column}")
        data_in_column = data_frame[column]
        unique_values = np.unique(data_in_column)
        print("Memory Usage:"+str(data_in_column.memory_usage()*0.001))
        num_values = len(unique_values)
        print(f"Number of unique values:{num_values}")
        if num_values<=50:
            print(f"The unique values:{unique_values}")
        print("")





def print_model_performance_for_classification(model, 
                                               features, 
                                               target, 
                                               threshold=DEFAULT_CLASSIFICATION_THRESHOLD):
    """
    Print the relevant scores for a classification model

    model: the model to classify with
    features: the feature vectors
    target: our target variable to predict
    threshold: threshold to infer as belonging to lable 1.
    """
    # predicting using the independent variables
    prediction = model.predict_proba(features)[:, 1]
    prediction_threshold = prediction > threshold
    final_prediction = np.round(prediction_threshold)
    # get the scores
    accuracy = accuracy_score(target, final_prediction)  
    recall = recall_score(target, final_prediction)  
    precision = precision_score(target, final_prediction)  
    f1 = f1_score(target, final_prediction)  
    
    print(f"Accuracy:{accuracy}")
    print(f"Recall:{recall}")
    print(f"Precision: {precision}")
    print(f"F1 score: {f1}")



def tune_model(model,
               parameters,
               X_train:pd.core.frame.DataFrame,
               y_train:pd.core.series.Series,):
    """Takes in model and parameters and return the tuned model.
    
    Parameters
    ----------
    model : any classifier 
    parameters : the parameters
    
    
    Return
    ------
    prunned_tree : prunned_tree with the tuned parameters
    """
    
    # we care about the recall score more, hence setting that
    acc_scorer = make_scorer(metrics.recall_score)

    # Run the grid search
    grid_obj = GridSearchCV(model, parameters, scoring=acc_scorer,cv=5)
    grid_obj = grid_obj.fit(X_train, y_train)

    # get the tuned model
    estimator_tuned = grid_obj.best_estimator_

    # Fir the data to the tuned model
    estimator_tuned.fit(X_train, y_train)
    
    return estimator_tuned

    
def box_plot_num_columns(columns):
    """For the provided columns do a box plot.
    
    Parameters
    ----------
    columns : column names
    """
    plt.figure(figsize=(30, 30))

    for idx, variable in enumerate(columns):
        plt.subplot(5, 5, idx + 1)
        sns.boxplot(data=data[variable])
        plt.tight_layout()
        plt.title(variable)

    plt.show()

def treat_outliers(df, col):
    """
    Treats outliers in a variable

    df: dataframe
    col: dataframe column
    """
    Q1 = df[col].quantile(0.25) 
    Q3 = df[col].quantile(0.75)  
    IQR = Q3 - Q1
    Lower_Whisker = Q1 - 1.5 * IQR
    Upper_Whisker = Q3 + 1.5 * IQR

   
    df[col] = np.clip(df[col], Lower_Whisker, Upper_Whisker)

    return df


def treat_all_outliers(df, num_cols):
    """
    Treat outliers in a list of variables

    df: dataframe
    col_list: list of dataframe columns
    """
    for c in num_cols:
        df = treat_outliers(df, c)
    return df


def get_metrics_score(model,
                      X_train,
                      X_test,
                      y_train,
                      y_test):
    '''
    Return all the metrics score of a model on training/testing
    
    Parameters
    ----------
    model : classifier to predict values of X

    '''
    
    scores=[] 
    
    #Predict on train and tests
    pred_train = model.predict(X_train)
    pred_test = model.predict(X_test)
    
    #Accuracy of the model
    train_acc = model.score(X_train,y_train)
    test_acc = model.score(X_test,y_test)
    
    #Recall of the model
    train_recall = metrics.recall_score(y_train,pred_train)
    test_recall = metrics.recall_score(y_test,pred_test)
    
    #Precision of the model
    train_precision = metrics.precision_score(y_train,pred_train)
    test_precision = metrics.precision_score(y_test,pred_test)
    
    scores.extend((train_acc,test_acc,train_recall,test_recall,train_precision,test_precision))
        
    
    return scores


def over_sample(x_train,y_train):
    """Oversample and return the data"""
    
    print("Before Oversampling, counts of label 'Yes': {}".format(sum(y_train == 1)))
    print("Before Oversampling, counts of label 'No': {} \n".format(sum(y_train == 0)))

    sm = SMOTE(
        sampling_strategy=1, k_neighbors=5, random_state=1
    )  
    x_train_over, y_train_over = sm.fit_resample(x_train, y_train)


    print("After Oversampling, counts of label 'Yes': {}".format(sum(y_train_over == 1)))
    print("After Oversampling, counts of label 'No': {} \n".format(sum(y_train_over == 0)))


    print("After Oversampling, the shape of train_X: {}".format(x_train_over.shape))
    print("After Oversampling, the shape of train_y: {} \n".format(y_train_over.shape))
    
    return x_train_over,y_train_over

def under_sample(x_train,y_train):
    """Undersample and return the data"""
    
    print("Before Undersampling, counts of label 'Yes': {}".format(sum(y_train == 1)))
    print("Before Undersampling, counts of label 'No': {} \n".format(sum(y_train == 0)))

    under_sample = RandomUnderSampler(random_state=1)
    x_train_un, y_train_un = under_sample.fit_resample(x_train, y_train)


    print("After Undersampling, counts of label 'Yes': {}".format(sum(y_train_un == 1)))
    print("After Undersampling, counts of label 'No': {} \n".format(sum(y_train_un == 0)))


    print("After Undersampling, the shape of train_X: {}".format(x_train_un.shape))
    print("After Undersampling, the shape of train_y: {} \n".format(y_train_un.shape))
    
    return x_train_un,y_train_un

def tune_with_rcv(model,x_train,y_train):
    """
     Takes in model and tunes it using RandomizedSearchCV and returns the param.
    """
    # Parameter grid to pass in RandomizedSearchCV
    param_grid_1={'n_estimators':np.arange(50,150,50),
                'scale_pos_weight':[2,5],
                'learning_rate':[0.01,0.1,0.2],
                'gamma':[0,1,3],
                'subsample':[0.8,1],
                'max_depth':np.arange(1,5,1),
                'reg_lambda':[5,10]
                 }
    
    param_grid_2 = {'n_estimators':np.arange(50,150,50),
                'learning_rate':[0.01,0.1,0.2],
                'subsample':[0.8,1],
                'max_depth':np.arange(1,5,1),
               }
    
    # Based on the model provided, choose the tuning parameters accordingly.
    if isinstance(model,GradientBoostingClassifier):
        param_grid=param_grid_2
    else:
        param_grid=param_grid_1
    
    # we care about the recall score more, hence setting that
    scorer = metrics.make_scorer(metrics.recall_score)
    
    tuned = RandomizedSearchCV(estimator=model, 
                                param_distributions=param_grid, 
                                n_iter=50, 
                                scoring=scorer, 
                                cv=5, 
                                random_state=1, 
                                n_jobs = -1)

    tuned.fit(x_train,y_train)

    print(f"Best parameters are {tuned.best_params_} with CV score={tuned.best_score_}")
    # return the tuned params and the score.
    return tuned.best_params_,tuned.best_score_
```

# Fixing the Income Category column

2. This is a typo for sure and it needs to be fixed and it is done below.

```python
def fix_income_category(val):
    if val == 'abc':
        return mode_income_category
    else:
        return val
# Fixing the income category
mode_income_category = data["Income_Category"].mode()[0]
data["Income_Category"] = data["Income_Category"].apply(lambda x: fix_income_category(x))
```

# which metric is right for model performance evaluation ?
Before we arrive on our metric, lets chart the main info down:
- True Positive: Person who is predicted going to churn has churned.
- True Negative: Person who is predicted not going to churn did not churn.
- False Positive: Person who is predicted to churn has not churned.
- False Negative: Person who is predicted as not churned, has churned.

So looking at **false negative**, that kind of incident will be of more harm to the company (since we are loosing a customer). Hence **Recall** should be our metric here.

- Please use **XGB_RC_TUNED** to predict if a customer will attrite or not. 
- Only take necessary reactions if you get >50% score.
- In the mean time, I will continue on improving the model.

## Qualities of customer who closes the account
- Most of the attrited customers earn less than 40 k.
- They are mostly users of Blue card.
- They have lesser credit limit.
- They have lesser utilization ratio (less than 0.2)
- They have more than 2.5 months of inactive status.
- Their total transaction amount is less than 3000.
- They is more contacts between them and the bank compared to existing customers.
- They have poor revolving balance.
- They use 1 or very less of our products.

 
## What can be done to avoid attrition of a customer?
- Often attrited customers have less utilization. To encourage them to use more, send promotion emails of our 
  cashback deals etc to them.
- Also attrited customers have more contact with the bank. So when they reach out, inform them about the new
  benefits in using the card.
- Increase the credit limit (lets start with 5% for now). The more spending power they have, the more interested they are to use the card.
- If there is a customer with just 1 month of inactive status, please reach out to them immediately with our benefits or offers. Because if left as is, they will probably attrite after a month.
- On a monthly basis, send promotion emails of our other products. The more products they have, less likely they are to leave.
