# WHAT IS NLP?
The class of techniques used to process and understand various aspects of unstructured text data, is called Natural Language Processing (NLP).

# Why?
We have studied, extracted information (EDA), and built models using numerical and image data so far. However, the vast majority of data in the world today is text within unstructured data - which constitutes up to 80% of the total data generated in the world these days. The rise in data and computing power has driven tremendous developments in the machine learning space, and the need to access and process text data is turning out to be increasingly significant with regards to solving real-life/business problems

# Some applications
 Some of the products/applications of NLP in life and business are Chatbots, Virtual assistants (Ex: Siri, Alexa), Text classification/categorization, Sentiment analysis, Machine translation, Speech recognition, Text summarization and Autocorrect. NLP applications are also used to analyze customer behavior (Customer analytics) and derive marketing intelligence.

- Vocabulary: A list of words used in the corpus.

# Stemming
- Convert all the extraploated forms of the word (for example ran, running runs will be changed to its base "run").

- can be done by removing sufixes and prefixes.
- advantage is reducing the dims in the text.
- but it may resut in words which is not even in dictionary (because it does not check). 

# Lemmatization
- very similar to stemming.
- ensures that the resultant word is part of the dictionary.
- it is time consuming, hence if not crtitical, people ususally go for stemming
