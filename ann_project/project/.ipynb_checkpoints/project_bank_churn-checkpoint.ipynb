```python
def print_frame_meta_info(data_frame:pd.DataFrame): 
    """Load a data frame file and print meta info of the data frame.

    Parameters
    ----------
    data_frame : DataFrame
    """
    rows,columns = np.shape(data_frame)
    print(f"Shape of the data frame is: {rows},{columns}")
    print(f"Number of variables are: {columns}")
    print("Memory Usage:"+str(data_frame.memory_usage(index=False,deep=True).sum()*0.001))
    print("")
    print("")

    
    for column in data_frame.columns:
        print(f"Name of the column: {column}")
        data_in_column = data_frame[column]
        unique_values = np.unique(data_in_column)
        print("Memory Usage:"+str(data_in_column.memory_usage()*0.001))
        num_values = len(unique_values)
        print(f"Number of unique values:{num_values}")
        if num_values<=50:
            print(f"The unique values:{unique_values}")
        print("")

def make_count_plot(col_name:str,
                    data:pd.DataFrame):
    """Make a count plot (and print info of each unique feature) for a feature in 
    a dataset."""
    fig, ax = plt.subplots()
    fig.set_size_inches(10.7, 10.27)
    ax = sns.countplot(x=col_name, data=data)
    ax.set_xticklabels(ax.get_xticklabels())
    plt.tight_layout()
    plt.show()
    
    print("")
    for status in data[col_name].unique():
        print(f"Number of customers who has {status} {col_name}: {len(data[(data.get(col_name) == status)])}")

def make_distribution_plot(col_name:str,
                            data:pd.DataFrame):
    """Make distribution plot (along with some diagnostics) for the given column 
    name and data frame"""
    print(f"Univariate Analysis of {col_name}")
    data_points = data.get(col_name)
    sns.displot(data_points,kde=True);
    print(f"Max {col_name} is {np.max(data_points)}")
    print(f"Min {col_name} is {np.min(data_points)}")
    print(f"Mean {col_name} is {np.mean(data_points)}")

def make_heat_map(data:pd.DataFrame):
  """Make a heatmap for a given dataset."""
  fig, ax = plt.subplots(figsize=(12,12))
  sns.set(font_scale = 2)
  sns.heatmap(data.corr(), 
              annot=True,
              ax=ax ,
              cbar=1, 
              square=1, 
              annot_kws={'size': 15}, 
              cmap= 'coolwarm')

def draw_bar_plot(x,y):
    """
    Draw bar plot for any 2 features of the table

    x: feature x
    y: feature y
    """
    fig, ax = plt.subplots()
    fig.set_size_inches(15.7, 5.27)
    ax = sns.barplot(x=x,y=y)
    plt.tight_layout()

def plot_train_vs_val(model_history):
    """Plot train vs validation loss for a model"""
    
    hist  = pd.DataFrame(model_history.history)
    hist['epoch'] = model_history.epoch

    # Accuracy plot
    plt.plot(hist['loss'])
    plt.plot(hist['val_loss'])
    plt.legend(("train" , "valid") )

def estimate_performance_metric(y_test,y_predicted):
    """Estimate the recall,precision and f1_score from the confusion matrix"""
    cf_matrix=confusion_matrix(y_test, y_predicted)
    
    recall    = cf_matrix[1,1] / sum(cf_matrix[1,:])
    precision = cf_matrix[1,1] / sum(cf_matrix[:,1])
    f1_score  = 2*precision*recall / (precision + recall)
    return recall,precision,f1_score
```

# Data Pre-processing - prepping the categorical features


```python
oneHotCols=["Geography","Gender"]
data_prepped=pd.get_dummies(data, columns=oneHotCols)
data_prepped.head(10)
```

# Data Pre-processing - Split the target variable and predictors


```python
X = data_prepped.drop(['Exited'],axis=1)
Y = data_prepped[['Exited']]
```

# Data Pre-processing - Split the data into train and test


```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, random_state = 0)
```

# Data Pre-processing - Rescale the data

```python
sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Model building - Initialize the parameters first

```python
# Get the number of input features for the input layer
_,num_input_features=np.shape(X_train)

# Get the number of output layers
_,num_output_layers=np.shape(y_train)

# Get the type of activation function for dense layers
dense_activation_type="relu"

# Get the type of activation function for dense layers
output_activation_type="sigmoid"

# Specify the loss function type
loss_function_type="binary_crossentropy"

# Specify the metric used
metric_type="accuracy"

# Specify the type of optimizer
optimizer_type="adam"
```

# Model building
## Model - 1
- Lets start with mininmal layers and epoch

```python

model = keras.Sequential([
    keras.layers.Dense(6, 
                       input_shape=(num_input_features,), 
                       activation=dense_activation_type),
    keras.layers.Dense(6, activation=dense_activation_type),
    keras.layers.Dense(num_output_layers,activation=output_activation_type)
])
model.compile(optimizer=optimizer_type,
              loss=loss_function_type,
              metrics=[metric_type])
history=model.fit(X_train, y_train, epochs=20,validation_split=0.2)
```

# Training vs validation plot (shown below) for model-1 above

```python
plot_train_vs_val(history)
```

# Recall score with model-1

```python
threshold=0.5
y_pred = model.predict(X_test)
for i in range(len(y_test)):
    if y_pred[i]>=0.5:
        y_pred[i]=1 
    else:
        y_pred[i]=0
recall,_,_ = estimate_performance_metric(y_test,y_pred)
print(f"Recall_score:{recall}")
```

# Model Performance Improvement - Which metric is right for model performance evaluation and why?
Lets establish our labels first:
- True Positive: A person predicted as leaving the bank, did leave the bank.
- True Negative: A person predicated as not leaving, did not leave the bank.
- False Positive: A person predicted as leaving the bank, did not leave the bank.
- False Negative: A person predicted as not leaving the bank, has left the bank.

So of all these, **False Negative** is the most harmful, since the bank looses a customer. So we should increase our **recall** to keep the **False Negative** to the minimum.

- Add the batch normalization again and see if it can improve the previous model

```python
model = keras.Sequential([
    keras.layers.Dense(65, 
                       input_shape=(num_input_features,), 
                       kernel_initializer='he_normal',
                       activation=dense_activation_type),
    keras.layers.Dropout(0.5),                   
    keras.layers.BatchNormalization(),
    keras.layers.Dense(65, activation=dense_activation_type,kernel_initializer='he_normal'),
    keras.layers.Dropout(0.5),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(num_output_layers,activation=output_activation_type,kernel_initializer='he_normal')
])
model.compile(optimizer=optimizer_type,
              loss=loss_function_type,
              metrics=[metric_type])
history=model.fit(X_train, y_train, epochs=200,validation_split=0.2)
```

# Model Performance Evaluation with Model 7
- Based on the ROC curve, above, **we will go with threshold 0.25** and estimate the recall score.

```python
threshold=0.25
y_pred = model.predict(X_test)
for i in range(len(y_test)):
    if y_pred[i]>=threshold:
        y_pred[i]=1 
    else:
        y_pred[i]=0
recall,precision,f1_score=estimate_performance_metric(y_test,y_pred)

print(f"Recall_score:{recall}")
print(f"precision_score:{precision}")
print(f"f1_score:{f1_score}")
```

# Model Performance Evaluation - Discussion
- with 0.25 threshold, we get a recall score of ~0.75, which is not bad. 
- This model **can help a lot in reducing false negatives** (which is very important for business to retain customers).
- The F1 score is around ~ 0.6, this is not ideal, but in the interest of a better recall score, we should just live with this.
